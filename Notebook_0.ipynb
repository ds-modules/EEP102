```python
import collections
from pylab import *
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
from scipy import misc
```

# Section 1

Consider the function f(x)= a x**2 + b x + c. 

## Parameters

You can change these parameters.

```python
a = 2
b = 2
c = 2
```

## Figure (Don't change the code below!)

The Figure shows the function f(x). It's first (f'(x)) and second (f''(x)) derivatives.

```python
x = np.linspace(-10, 10, 20) 

y = a*(x**2)+ b*x + c

y2 = (2*a*x) + b

y3 = np.linspace(a, a, 20) 


fig = plt.figure()
ax = plt.subplot(111)
ax.plot(x, y,'r', label='f(x)')
ax.plot(x, y2,'b', label='f(x)')
ax.plot(x, y3,'g',label='f(x)')
plt.title('Derivatives')
ax.legend()
plt.show()
 
```

```python
ax = subplot(111)

def fonction(x):
    return a*x*x+b*x+c

x = arange(-2.0, 2.0, 0.01)

y = fonction(x)

plot(x, y,'r-')

yp = misc.derivative(fonction, x)

plot(x, yp,'b-')

ypp = misc.derivative(fonction, yp)

plot(x, ypp,'g-')

grid(True)

ax.spines['left'].set_position('zero')
ax.spines['right'].set_color('none')
ax.spines['bottom'].set_position('zero')
ax.spines['top'].set_color('none')

text(-0.75, 6.0,
     r'$f(x)$', horizontalalignment='center',
     fontsize=18,color='red')

text(-1.0, -8.0,
     r"$f'(x)$", horizontalalignment='center',
     fontsize=18,color='blue')

show()
```
